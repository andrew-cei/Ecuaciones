# Ecuación de decaimiento

Es una ecuación diferencial que modela el decaimiento radioactivo y tiene la siguiente forma:

$$\frac{dx}{dt}=kx,$$

donde $t$ es el tiempo y $x$ la concentración de la sustancia radioactiva, además de una condición inicial al tiempo cero $x(0)=x_0$.

Esta ecuación se puede resolver mediante separación de variables de la siguiente forma:

$$\frac{dx}{x}=kdt,$$
$$\int \frac{dx}{x}=\int kdt,$$
$$Ln(x)=kt + C,$$
$$x(t) = Ae^{kt}.$$

Al aplicar la condición inicial se obtiene:

$$x(t)=x_0e^{kt}.$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Ejemplo práctico

Consideremos $k=-2$ y la condición inicial $x_0=10$, se tiene el siguiente problema:

$$\frac{dx}{dt}=-2x,$$

cuya solución es:

$$x(t) = 10 e^{-2t}.$$

Una gráfica de la solución en el intervalo $[0,2]$ es la siguiente:

```python
# Solución exacta
t = np.linspace(0,1,1000)
x = 10*np.exp(-2*t)
plt.plot(t,x,linewidth = 2)
plt.grid()
```

```python
# Solución aproximada con el método de Euler
# Ecuación de decaimiento radioactivo con constante k = -2
# dx = -2x x(0)=10
def f(t,x):
    return -2*x

# Método de Euler
# Límites del intervalo, paso y condición inicial
a = 0
b = 1
h = 0.2
x_0 = 10

# Arreglo de coordenadas x y coordenadas y con condición inicial
t_lst = [a + i * h for i in range(int((b - a) / h) + 1)]
x_lst = []

# Implementación del método
for ti in t_lst:
    if(len(x_lst) == 0):
        x_lst.append(x_0)
    else:
        x_lst.append(x_lst[-1] + h*f(ti,x_lst[-1]))
# Gráfica
plt.plot(t_lst,x_lst,linewidth = 2)
plt.grid()
```

```python
# Comparación
plt.plot(t, x, label='Exacta', color='blue') # Plot the first line
plt.plot(t_lst, x_lst, label='Euler', color='red', linestyle='--') # Plot the second line
plt.legend() # Show a legend to differentiate the lines
plt.grid()
```

## Conclusiones

A partir de la gráfica anterior, se concluye que el método de Euler comete un error grande al tomar el paso de $h=0.2$ para un intervalo $[0,1]$, por lo que se recomienda disminuir su tamaño.
